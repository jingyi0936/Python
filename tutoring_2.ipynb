```python
for fahr in [30, 40, 50, 60, 70]:
  # Below are loop body
  cent = (fahr - 32) / 9.0 * 5
  print(fahr, cent)
  print("great") # This will be printed five times, because it is inside the loop

print("y") # This will be only printed one time, because it is outside the loop
```
